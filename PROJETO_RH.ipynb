```python
#As pessoas estão me falando que o pessoal de Vendas está pedindo demissão. Você consegue confirmar isso com dados? Quero saber qual departamento tem a MAIOR proporção (%) de desligados em relação ao total de funcionários daquele departamento — não em número absoluto, em percentual. E me diz também qual é o principal motivo de desligamento em Vendas
resultado = df.groupby("departamento").agg(
    demissao = ("status", lambda x: (x=="Desligado").sum()),
    total = ("status", "size")
)
resultado["porcertagem"] = resultado["demissao"] / resultado["total"]
resultado["porcertagem"]
```

```python
#Queria comparar a satisfação média de quem faz hora extra com a de quem não faz. Se possível, quebra isso por departamento também, porque acho que no meu setor é diferente do resto
satisfação_dep = df.groupby(["horas_extras","departamento"])["satisfacao"].mean().unstack()
```

```python
#Separa os funcionários ativos do meu departamento em faixas: até 1 ano de casa, de 1 a 3 anos, e mais de 3 anos. Quero saber quantos tem em cada faixa
operacao_ativos  = df.loc[(df["departamento"] == "Operações") & (df["status"] == "Ativo")].copy()
operacao_ativos["tempo_empresa"] = pd.cut(operacao_ativos["tempo_empresa_meses"], bins= [- np.inf,12,36,df['tempo_empresa_meses'].max()], labels=["Até 1 ano",
        "De 1 a 3 anos",
        "Mais de 3 anos"])
operacao_ativos["tempo_empresa"]
```
